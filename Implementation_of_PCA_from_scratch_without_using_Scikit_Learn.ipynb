```python
from sklearn.decomposition import PCA
import pandas as pd
import numpy as np
```

```python
training_data = pd.read_csv("/content/sample_data/california_housing_train.csv")
```

```python
training_data
```

```python
Y = training_data.drop(labels=training_data.columns[-1],axis=1)
```

```python
Y
```

```python
pca = PCA()
```

```python
projected_X = pca.fit_transform(Y)
```

```python
projected_X
```

```python
projected_X = pd.DataFrame(data=projected_X,
                           columns=["Feature_"+str(i) for i in range(1,9)])
```

```python
projected_X
```

```python
pca.singular_values_
```

```python
pca.get_covariance()
```

```python
Y_zero_mean = Y - Y.mean()
```

```python
Y_zero_mean = np.array(Y_zero_mean)
```

```python
sigma_original = (1/Y.shape[0])*np.matmul(Y_zero_mean.T,
                                                   Y_zero_mean)
```

```python
sigma_original_direct = Y.cov()
```

```python
sigma_original_direct
```

```python
sigma_original
```

```python
factorized_matrices = np.linalg.svd(sigma_original)
```

```python
factorized_matrices
```

```python
projected_X = np.matmul(Y_zero_mean,factorized_matrices[0])
```

```python
projected_X.shape
```

```python
projected_X = pd.DataFrame(data=projected_X,columns=["Feature_"+str(i) for i in range(1,9)])
```

```python
projected_X
```

```python
projected_X.cov()
```

```python
n_components = 6
```

```python
Q = factorized_matrices[0]
```

```python
Q_tilda = Q[:,0:n_components]
```

```python
projected_X_dim_reduced = np.matmul(Y_zero_mean,Q_tilda)
```

```python
projected_X_dim_reduced = pd.DataFrame(data=projected_X_dim_reduced,
                                       columns=["Feature_"+str(i) for i in range(1,7)])
```
